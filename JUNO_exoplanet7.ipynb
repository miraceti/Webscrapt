```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://exoplanet.eu/catalog/#catalog-table-section'
page = requests.get(url)
soup = BeautifulSoup(page.text,features="html")
print('soup\n',soup)
```

```python
soup1 =soup.find_all('section', id='catalog-table-section')
soup1
```

```python
soup1[0]
```

```python
soup2 = soup1[0].find_all('div')
soup2
```

```python
soup2 = soup1[0].find_all('div')
soup2[13]
```

```python
soup3 = soup2[13].find_all('div')[1]
soup3
```

```python
soup3_titres=soup3.find_all('th')
soup3_tables_titres = [titre.text.strip() for titre in soup3_titres ]
print('SOUP3TT\n',soup3_tables_titres)
```

```python
df = pd.DataFrame(columns=soup3_tables_titres)
df
```
