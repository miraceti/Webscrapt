```python
import pandas as pd
import flet as ft
from urllib.request import urlopen
import json
```

```python
urlexo1 = "https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+count(pl_name)+as+nbe+from+ps+where+default_flag=1&format=json"
print(urlexo1)
data = json.loads(urlopen(urlexo1).read().decode("utf-8"))
# print(data)
data0=data[0]
# print(data0)
nb_exoplanets= data0['nbe']
print(nb_exoplanets)
```

```python
urlexo7 = "https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+\
pl_name,hostname,pl_letter,pl_bmasse,pl_bmassj,pl_rade,pl_radj,pl_orbper,pl_dens,pl_trandur,pl_ratror,\
sy_snum,sy_pnum,sy_mnum,st_mass,st_lum,st_age,st_dens,sy_dist,disc_year,disc_telescope,discoverymethod\
+from+ps+where+default_flag=1&format=json"

print(urlexo7)
list7 = json.loads(urlopen(urlexo7).read().decode("utf-8"))
#print(list7)
```

```python
#list7
```

```python
df_exo7 = pd.DataFrame.from_records(list7)
df_exo7
```

```python
df_exo7.columns
```

```python
df_exo7['sy_dist']#en parsecs=3.26 annee lumiere
```

```python
df_exo7[df_exo7['sy_dist'] < 10.00]
```

```python
nb_df_exo7_sy_dist_inf_10=df_exo7[df_exo7['sy_dist'] < 10.00]['pl_name'].count()
nb_df_exo7_sy_dist_inf_20=df_exo7[(df_exo7['sy_dist'] >= 10.00)&(df_exo7['sy_dist'] < 20.00)]['pl_name'].count()
nb_df_exo7_sy_dist_inf_30=df_exo7[(df_exo7['sy_dist'] >= 20.00)&(df_exo7['sy_dist'] < 30.00)]['pl_name'].count()

print(nb_df_exo7_sy_dist_inf_10)
print(nb_df_exo7_sy_dist_inf_20)
print(nb_df_exo7_sy_dist_inf_30)
```
